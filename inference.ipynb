```python
import torch
device = torch.device("cuda")
```

```python
from transformers import AutoTokenizer, ElectraForSequenceClassification

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("monologg/koelectra-base-v3-discriminator")
model = ElectraForSequenceClassification.from_pretrained("monologg/koelectra-base-v3-discriminator", num_labels=4)
model.to(device)
model.load_state_dict(torch.load("/content/drive/MyDrive/model.pt"))
```

```python
# Input text
input_text = ""

# Tokenize input text
inputs = tokenizer(
        input_text, return_tensors="pt",
        truncation=True,
        max_length=256,
        pad_to_max_length=True,
        add_special_tokens=True)

# Forward pass through the model
with torch.no_grad():
    outputs = model(**inputs)

# Get predicted probabilities
probabilities = torch.softmax(outputs.logits, dim=1)

threshold = 0.3
max_probabilities, predicted_classes = torch.max(probabilities, dim=1)

# Mark instances as unclassified (label 0) if max probability is below the threshold
predicted_classes[max_probabilities < threshold] = 0

# Print results
print("Predicted class:", predicted_classes.item())
print("Probabilities:", probabilities.tolist())
```
